```python
import pandas as pd
import matplotlib as plt
```

```python
file_path = "../50TH-Percentile-Rent-Estimates-Data/Merged_Data/rent20-22.csv"
DFrame = pd.read_csv(file_path)
```

```python
DFrame.head()
```

```python
pruneDF = DFrame
del pruneDF["pop2017"]
del pruneDF["Unnamed: 0"]
pruneDF.head()
```

```python
twenty_studio_max = pruneDF["studio_2020"].max()
twenty_studio_min = pruneDF["studio_2020"].min()
twentyone_studio_max = pruneDF["studio_2021"].max()
twentyone_studio_min = pruneDF["studio_2021"].min()
twentytwo_studio_max = pruneDF["studio_2022"].max()
twentytwo_studio_min = pruneDF["studio_2022"].min()
print(f"For Studios, the maximum rent price in 2020 is ${twenty_studio_max}, and the minimum is ${twenty_studio_min}. The maximum rent price in 2021 is ${twentyone_studio_max}, and the minimum is ${twentyone_studio_min}. Lastly, the maximum rent price in 2022 is ${twentytwo_studio_max}, and the minimum is ${twentytwo_studio_min}.")
```

```python
twenty_onebed_max = pruneDF["bedroom_1_2020"].max()
twenty_onebed_min = pruneDF["bedroom_1_2020"].min()
twentyone_onebed_max = pruneDF["bedroom_1_2021"].max()
twentyone_onebed_min = pruneDF["bedroom_1_2021"].min()
twentytwo_onebed_max = pruneDF["bedroom_1_2022"].max()
twentytwo_onebed_min = pruneDF["bedroom_1_2022"].min()
print(f"For one-bedrooms, the maximum rent price in 2020 is ${twenty_onebed_max}, and the minimum is ${twenty_onebed_min}. The maximum rent price in 2021 is ${twentyone_onebed_max}, and the minimum is ${twentyone_onebed_min}. Lastly, the maximum rent price in 2022 is ${twentytwo_onebed_max}, and the minimum is ${twentytwo_onebed_min}.")
```

```python
twenty_twobed_max = pruneDF["bedroom_2_2020"].max()
twenty_twobed_min = pruneDF["bedroom_2_2020"].min()
twentyone_twobed_max = pruneDF["bedroom_2_2021"].max()
twentyone_twobed_min = pruneDF["bedroom_2_2021"].min()
twentytwo_twobed_max = pruneDF["bedroom_2_2022"].max()
twentytwo_twobed_min = pruneDF["bedroom_2_2022"].min()
print(f"For two-bedrooms, the maximum rent price in 2020 is ${twenty_twobed_max}, and the minimum is ${twenty_twobed_min}. The maximum rent price in 2021 is ${twentyone_twobed_max}, and the minimum is ${twentyone_twobed_min}. Lastly, the maximum rent price in 2022 is ${twentytwo_twobed_max}, and the minimum is ${twentytwo_twobed_min}.")
```

```python
twenty_threebed_max = pruneDF["bedroom_3_2020"].max()
twenty_threebed_min = pruneDF["bedroom_3_2020"].min()
twentyone_threebed_max = pruneDF["bedroom_3_2021"].max()
twentyone_threebed_min = pruneDF["bedroom_3_2021"].min()
twentytwo_threebed_max = pruneDF["bedroom_3_2022"].max()
twentytwo_threebed_min = pruneDF["bedroom_3_2022"].min()
print(f"For three-bedrooms, the maximum rent price in 2020 is ${twenty_threebed_max}, and the minimum is ${twenty_threebed_min}. The maximum rent price in 2021 is ${twentyone_threebed_max}, and the minimum is ${twentyone_threebed_min}. Lastly, the maximum rent price in 2022 is ${twentytwo_threebed_max}, and the minimum is ${twentytwo_threebed_min}.")
```

```python
twenty_fourbed_max = pruneDF["bedroom_4_2020"].max()
twenty_fourbed_min = pruneDF["bedroom_4_2020"].min()
twentyone_fourbed_max = pruneDF["bedroom_4_2021"].max()
twentyone_fourbed_min = pruneDF["bedroom_4_2021"].min()
twentytwo_fourbed_max = pruneDF["bedroom_4_2022"].max()
twentytwo_fourbed_min = pruneDF["bedroom_4_2022"].min()
print(f"For three-bedrooms, the maximum rent price in 2020 is ${twenty_fourbed_max}, and the minimum is ${twenty_fourbed_min}. The maximum rent price in 2021 is ${twentyone_fourbed_max}, and the minimum is ${twentyone_fourbed_min}. Lastly, the maximum rent price in 2022 is ${twentytwo_fourbed_max}, and the minimum is ${twentytwo_fourbed_min}.")
```
